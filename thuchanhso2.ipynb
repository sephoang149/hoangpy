```python
#bài 15
a='Xin chào, tôi là Python. Python là ngôn ngữ lập trình!'
dau='.!?,'
A=''
for i in a :
    if i not in dau:
        A+=i
print(A)
b=A.split()
c=list(set(b))
d=[b.count(tu) for tu in c]
tu_dem = list(zip(c, d))
tu_dem.sort(key=lambda x: x[1], reverse=True)
for tu, so_lan in tu_dem:
    print(f"{tu} xuất hiện {so_lan} lần")
```
